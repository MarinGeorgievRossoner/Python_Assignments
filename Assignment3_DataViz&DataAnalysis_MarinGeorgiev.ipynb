```python
#Which of the following best describes you: During the pandemic I have...
#Ended a romantic relationship - 3
#Started a romantic relationship - 2
#Remained in a romantic relationship (living with my partner or saw him/her regularly) - 23
#Remained single - 11
#Multiple of the above - 5
#None of the above - 1

ended_rr = 3
started_rr = 2
remained_rr = 23
remained_single = 11
multiple_rr = 5 
no_rr = 1

total_rr = ended_rr + started_rr + remained_rr + remained_single + multiple_rr + no_rr
print("The total number of answers is:",total_rr)

perc_ended_rr = ended_rr*100/45
perc_started_rr = started_rr*100/45
perc_remained_rr = remained_rr*100/45
perc_remained_single = remained_single*100/45
perc_multiple_rr = multiple_rr*100/45
perc_no_rr = no_rr*100/45

print("Percentage of people who ended a RR: ","{:.2f}".format(perc_ended_rr))
print("Percentage of people who started a RR: ","{:.2f}".format(perc_started_rr))
print("Percentage of people who remained in a RR: ","{:.2f}".format(perc_remained_rr))
print("Percentage of people who remained  single: ","{:.2f}".format(perc_remained_single))
print("Percentage of people who had multiple of the above: ","{:.2f}".format(perc_multiple_rr))
print("Percentage of people who had nothing of the above: ","{:.2f}".format(perc_no_rr))
```

```python
#Comparison of online and face-to-face contacts before and during the pandemics.
#Social groups compared: parents, siblings, grandparents, friends in the same city, friends abroad,
#strangers, co-workers, other relatives.

#Daily, Weekly, Monthly, Yearly, and No face-to-face contact before and during the pandemic.

daily_face_before = 79
weekly_face_before = 118 
monthly_face_before = 96
yearly_face_before = 89
no_face_before = 66

daily_face_during = 28
weekly_face_during = 52
monthly_face_during = 88
yearly_face_during = 106
no_face_during = 181


daily_face_change = (daily_face_during - daily_face_before)*100/daily_face_before
weekly_face_change = (weekly_face_during - weekly_face_before)*100/weekly_face_before
monthly_face_change = (monthly_face_during - monthly_face_before)*100/monthly_face_before
yearly_face_change = (yearly_face_during - yearly_face_before)*100/yearly_face_before
no_face_change = (no_face_during - no_face_before)*100/no_face_before

print("Change in daily face-to-face contact: ","{:.2f}".format(daily_face_change))
print("Change in weekly face-to-face contact: ","{:.2f}".format(weekly_face_change))
print("Change in monthly face-to-face contact: ","{:.2f}".format(monthly_face_change))
print("Change in yearly face-to-face contact: ","{:.2f}".format(yearly_face_change))
print("Change in no face-to-face contact: ","{:.2f}".format(no_face_change),"\n")

#Daily, Weekly, Monthly, Yearly, and No online contact before and during the pandemic.

daily_online_before = 33
weekly_online_before = 89 
monthly_online_before = 91
yearly_online_before = 59
no_online_before = 168

daily_online_during = 51
weekly_online_during = 125
monthly_online_during = 97
yearly_online_during = 43
no_online_during = 135

daily_online_change = (daily_online_during - daily_online_before)*100/daily_online_before
weekly_online_change = (weekly_online_during - weekly_online_before)*100/weekly_online_before
monthly_online_change = (monthly_online_during - monthly_online_before)*100/monthly_online_before
yearly_online_change = (yearly_online_during - yearly_online_before)*100/yearly_online_before
no_online_change = (no_online_during - no_online_before)*100/no_online_before

print("Change in daily online contact: ","{:.2f}".format(daily_online_change))
print("Change in weekly online contact: ","{:.2f}".format(weekly_online_change))
print("Change in monthly online contact: ","{:.2f}".format(monthly_online_change))
print("Change in yearly online contact: ","{:.2f}".format(yearly_online_change))
print("Change in no online contact: ","{:.2f}".format(no_online_change))
```
